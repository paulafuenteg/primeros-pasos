- Cambiad el nombre de las columnas para que todas estén en minúsculas y reemplazad los espacios por _.

```python
#Creamos un diccionario con los nombres de las columnas en minusculas y cambiando el espacio con _ mediante una list conprenhension
nuevas_columnas = {col:col.lower().replace(' ','_') for col in attacks.columns}
#Renombramos las columnas con el método rename
attacks.rename(columns = nuevas_columnas,inplace = True)
attacks.head(2)
```

- Eliminad las columnas de Unnamed: 22 y Unnamed: 23, Case Number.1, Case Number.2, original order, Investigator or Source, pdf, href formula.

```python
attacks.drop(['unnamed:_22','unnamed:_23','case_number.1','case_number.2','original_order','investigator_or_source','pdf','href_formula'],axis =1, inplace = True)
```
