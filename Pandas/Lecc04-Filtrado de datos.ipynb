8- Si extraemos los valores únicos de la columna species nos podemos dar cuenta que está un poco sucia y extraer informacion útil de la misma se puede convertir en algo difícil. En este ejercicio tendréis que extraer los ataques que han ocurrido por tiburón blanco (lo encontraremos en la columna species como algo que contenga white o White). Usad el método str.contains() para resolver este ejercicio.

```python
patron = '.+white.+' #Creamos el patrón de regex para cualquier dato que incluya la palabra white
```

```python
white = df[df['species_'].str.contains(patron, case = False, regex = True)] #filtramos el df con str.contains incluyendo el patron anterior y case = false para que no sea case sensitive y encuentre también White
white.head()
```

9- Extraed ahora los ataques por tiburón blanco o tigre. Utilizad el método str.contains() para resolver este ejercicio.

```python
patron1 ='.+white.+|.+tiger.+' #Creamos un patrón para que encuentre todo lo que contenga white o tiger
```

```python
white_tiger = df[df['species_'].str.contains(patron1, case = False, regex = True)] #Filtramos con el patron anterior
white_tiger.head()
```

10- Utilizando la columna de year cread una nueva columna llamada siglo que tiene siglo-XX o siglo-XXI. Donde el año sea menor que 2000 pondremos siglo-XX y el resto de los años siglo-XXI.

```python
condiciones = [df["year"] < 2000.0, df["year"] >= 2000.0] #Creamos los filtros de los años
opciones = ["siglo-XX", "siglo-XXI"] #Creamos las opciones que incluiran las columnas

df['siglo']= np.select(condiciones, opciones) #A través de np.select creamos la columna nueva
df.head()
```

```python
df['siglo'].value_counts() #Comprobamos los valores del df
```
