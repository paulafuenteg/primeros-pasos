8- Si extraemos los valores √∫nicos de la columna species nos podemos dar cuenta que est√° un poco sucia y extraer informacion √∫til de la misma se puede convertir en algo dif√≠cil. En este ejercicio tendr√©is que extraer los ataques que han ocurrido por tibur√≥n blanco (lo encontraremos en la columna species como algo que contenga white o White). Usad el m√©todo str.contains() para resolver este ejercicio.

```python
patron = '.+white.+' #Creamos el patr√≥n de regex para cualquier dato que incluya la palabra white
```

```python
white = df[df['species_'].str.contains(patron, case = False, regex = True)] #filtramos el df con str.contains incluyendo el patron anterior y case = false para que no sea case sensitive y encuentre tambi√©n White
white.head()
```

9- Extraed ahora los ataques por tibur√≥n blanco o tigre. Utilizad el m√©todo str.contains() para resolver este ejercicio.

```python
patron1 ='.+white.+|.+tiger.+' #Creamos un patr√≥n para que encuentre todo lo que contenga white o tiger
```

```python
white_tiger = df[df['species_'].str.contains(patron1, case = False, regex = True)] #Filtramos con el patron anterior
white_tiger.head()
```

10- Utilizando la columna de year cread una nueva columna llamada siglo que tiene siglo-XX o siglo-XXI. Donde el a√±o sea menor que 2000 pondremos siglo-XX y el resto de los a√±os siglo-XXI.

```python
condiciones = [df["year"] < 2000.0, df["year"] >= 2000.0] #Creamos los filtros de los a√±os
opciones = ["siglo-XX", "siglo-XXI"] #Creamos las opciones que incluiran las columnas

df['siglo']= np.select(condiciones, opciones) #A trav√©s de np.select creamos la columna nueva
df.head()
```

```python
df['siglo'].value_counts() #Comprobamos los valores del df
```
