```python
import pandas as pd
```

```python
#Leemos el archivo y lo guardamos en la variable pandas 5
attacks5 = pd.read_csv('Datos/attacks_pandas_5.csv')
attacks5.head(2)
```

1- Â¿CuÃ¡ntos ataques hubo por paÃ­s? Devuelve un DataFrame.

ðŸ“Œ NOTA Ordena los resultados de mayor a menor. Esto todavÃ­a no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). 


```python
#Creamos un df con la suma de los ataques por pais
pais = attacks5.groupby('country')['case_number'].agg(ataques=('count')) #Usamos agg para poder cambiarle el nombre a la columna 
pais = pais.reset_index()
pais.head()
```

```python
#Ordenamos el df por la columna ataques de mayor a menor
pais.sort_values(by='ataques', ascending = False, inplace = True)
pais.head()
```

4- Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. Â¿CuÃ¡ntos de los ataques fueron letales por sexo? Devuelve un DataFrame.


```python
#Creamos un df con los ataques ocurridos en USA, AUSTRALIA Y SUDAFRICA
usa_aus_sa = attacks5[attacks5['country'].isin(['USA','AUSTRALIA', 'SOUTH AFRICA'])]
usa_aus_sa.head(2)
```

```python
#Filtramos el dataframe por los accidentes que hayan sido fatales(fatal=Y)
fatal = usa_aus_sa[usa_aus_sa['fatal_(y/n)']=='Y']
fatal.head(2)
```

```python
#Agrupamos el df fatal por el sexo y contamos los casos por sexo
fatal_genero = fatal.groupby('sex_')['case_number'].agg(ataques=('count'))
fatal_genero= fatal_genero.reset_index()
fatal_genero
```
