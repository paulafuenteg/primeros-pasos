2. Chequead que los nombres de las columnas ver si se llaman iguales entre dataframes . ¿Están todas en minúsculas? ¿Tienen espacios? En caso de que no sean todas iguales unificad el nombre de las columnas.

```python
#Ponemos todas las columnas en minúsculas y remplazamos los espacios por _ con una list comprehension y renombramos a las columnas de ambos df
nuevas_columnas = {col:col.lower().replace(' ','_') for col in adicional.columns}
adicional.rename(columns = nuevas_columnas,inplace = True)
adicional.head(2)
```

```python
nuevas_columnas1 = {col:col.lower().replace(' ','_') for col in historico.columns}
historico.rename(columns = nuevas_columnas1,inplace = True)
historico.head(2)
```

3. En este momento nos habremos dado cuenta que el DataFrame que generamos en el pair programming Pandas IV es similar al DataFrame de los datos historicos. Si recordamos, eliminamos algunas columnas en la sesión III que no eran útiles, pero esas columnas están todavía en el histórico. Si no fuera por eso podríamos unir esos dataframes.
```python
#Eliminamos las columnas en historico
historico.drop(['unnamed:_22','unnamed:_23','case_number.1','case_number.2','original_order','investigator_or_source','pdf','href_formula'],axis =1, inplace = True)
```

```python
#Ponemos case_number como index de histórico
historico.set_index('case_number',inplace = True)
```

```python
attacks4.columns
```

```python
#Para que histórico tenga también la columna siglo, la creamos igual que en el pair anterior
condiciones = [historico["year"] < 2000.0, historico["year"] >= 2000.0] #Creamos los filtros de los años
opciones = ["siglo-XX", "siglo-XXI"] #Creamos las opciones que incluiran las columnas

historico['siglo']= np.select(condiciones, opciones) #A través de np.select creamos la columna nueva
historico.head()
```

```python
attacks4.columns
```

```python
#Unimos attacks con histórico, como tienen las mismas colúmnas los juntamos mediante concat con el axis = 0 para que se añadan las filas de histórico a attacks
concat_historico_attacks = pd.concat([historico,attacks4], axis = 0)
concat_historico_attacks.head()
```

```python
historico.size + attacks4.size == concat_historico_attacks.size 
# Comprobamos que se ha unido viendo que tienen el mismo size
```

4. Es el momento de unir la información de todos los ataques de tiburón que acabamos de crear con el DataFrame de información adicional.

```python
# Convertimos a índice la columna case_number para hacer el join
adicional.set_index('case_number',inplace = True) 
```

```python
#Unimos el df creado de concatenar histórico con attacks con adicional mediante un lef join teniendo en cuenta el indice case_number que sería el dato común
joins = concat_historico_attacks.join(adicional, on ='case_number') 
```
